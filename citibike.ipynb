```python
# Dependencies
import pandas as pd
```

```python
# List of file names to be combined
files = ['data/tripdata1.csv/JC-202212-citibike-tripdata.csv', 'data/tripdata2.csv/JC-202301-citibike-tripdata.csv', 'data/tripdata3.csv/JC-202302-citibike-tripdata.csv', 'data/tripdata4csv/JC-202303-citibike-tripdata.csv', 'data/tripdata5.csv/JC-202304-citibike-tripdata.csv']

# Create an empty list to store each dataframe
df_list = []

# Loop through each file in the list
for file in files:
    # Read the file into a dataframe
    df_files = pd.read_csv(file)

    # Append the dataframe to the list
    df_list.append(df_files)

# Concatenate all dataframes in the list into one dataframe
df_files = pd.concat(df_list)

# Show dataframe
df_files
```

```python
# Save the dataframe to a CSV file
df_files.to_csv('data/citibike.csv', index=False)
```
